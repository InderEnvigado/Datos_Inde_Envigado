**Limpieza de datos**

```python
##Renombramos las columnas
INSCRITOSIE.rename(columns={"fecha": "fecha_inscripción"})
```

```python
##Renombramos las columnas

INSCRITOSIE = INSCRITOSIE.rename(columns={"Inscripcion": "Nro_inscripcion"})
INSCRITOSIE = INSCRITOSIE.rename(columns={"nom_barrio": "Barrio"})
INSCRITOSIE = INSCRITOSIE.rename(columns={"Servicio": "Disciplina"})
INSCRITOSIE = INSCRITOSIE.rename(columns={"cod_ciu": "Municipio_residencia"})
INSCRITOSIE
```

```python
# Convertir la columna 'Disciplina' a texto en mayúsculas
INSCRITOSIE['Disciplina'] = INSCRITOSIE['Disciplina'].str.upper()
```

```python
## transformar el tipo de dato en algunas columnas.

INSCRITOSIE = INSCRITOSIE.astype({'Nro_inscripcion': 'int32'})

INSCRITOSIE['Nivel'] = INSCRITOSIE['Nivel'].astype('category')

INSCRITOSIE = INSCRITOSIE.astype({'Disciplina': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'Barrio': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'edad': 'int32'})

INSCRITOSIE = INSCRITOSIE.astype({'Género': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'etnia': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'Poblacion': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'Escenario': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'instructor': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'curso': 'int32'})

INSCRITOSIE = INSCRITOSIE.astype({'zona': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'eps': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'estrato': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'Municipio_residencia': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'tipo_documento': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'nacimiento': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'rh': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'nacionalidad': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'escolaridad': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'diversidad': 'category'})

INSCRITOSIE = INSCRITOSIE.astype({'Población': 'category'})




```

```python
# El metodo replace Identifica el valor a buscar y luego reemplaza en todos los campos con el valor deseado.
INSCRITOSIE = INSCRITOSIE.replace("", None)
INSCRITOSIE
```

```python
# Limpiamos columnas de texto eliminando espacios innecesarios y convertimos primeras letras de una palabra a mayúsculas
def limpiar_texto(texto):
    if isinstance(texto, str):  # Verificamos que el valor sea string
        return ' '.join(texto.strip().title().split())  # Elimina espacios innecesarios y convierte el formato
    return texto  # Si no es texto, se devuelve el valor original sin cambios

# Lista de columnas de texto que se van a limpiar
columnas_texto = ['fecha', 'Nro_inscripcion', 'Disciplina', 'Nivel', 'Barrio', 'edad',
                   'Género', 'etnia', 'Poblacion', 'Escenario', 'instructor',
                   'curso', 'zona', 'fecha_nacimiento', 'eps',
                   'estrato', 'Municipio_residencia', 'tipo_documento','nacimiento', 'rh','nacionalidad', 'escolaridad','diversidad','Población' ]

# Aplicación de la función de limpieza a las columnas de texto
INSCRITOSIE[columnas_texto] = INSCRITOSIE[columnas_texto].applymap(limpiar_texto)
```

```python
# Reemplazar los valores nulos por None y No APLICA por OTRO
INSCRITOSIE['Barrio'] = INSCRITOSIE['Barrio'].replace({'NULL': None})
INSCRITOSIE['Barrio'] = INSCRITOSIE['Barrio'].replace({'No Aplica': 'Otro'})
```

```python
# verificar valores únicos en Barrio
print(INSCRITOSIE['Barrio'].unique())
```

```python
## # verificar valores únicos en Disciplina
print(INSCRITOSIE['Disciplina'].unique())
```

```python
## Esta opción te mostrará los valores únicos junto con su frecuencia, ordenados de mayor a menor:
## Usar prettytable para una visualización tabular
from prettytable import PrettyTable

# Contar los valores únicos y ordenarlos
disciplina_counts = INSCRITOSIE['Disciplina'].value_counts().sort_values(ascending=False)

# Crear una tabla
table = PrettyTable()
table.field_names = ["Disciplina", "Frecuencia"]

# Agregar los valores únicos a la tabla
for disciplina, frecuencia in disciplina_counts.items():
    table.add_row([disciplina, frecuencia])

# Imprimir la tabla
print(table)
```

```python
## Usa matplotlib o seaborn para graficar.
import pandas as pd
import matplotlib.pyplot as plt

# Contar y ordenar las disciplinas
disciplina_counts = INSCRITOSIE['Disciplina'].value_counts().sort_values(ascending=False)

# Crear un gráfico de barras horizontales
plt.figure(figsize=(12, 8))
disciplina_counts.plot(kind='barh', color='lightgreen')
plt.title('Frecuencia de Disciplinas (Ordenado de Mayor a Menor)', fontsize=16)
plt.xlabel('Frecuencia', fontsize=14)
plt.ylabel('Disciplina', fontsize=14)
plt.tight_layout()  # Ajustar el layout
plt.show()
```
